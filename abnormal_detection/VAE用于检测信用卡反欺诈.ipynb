```python
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.backend as K
from sklearn.cluster import KMeans
```

```python
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], trainset.shape[1]), mean=0, stddev=1)
    return z_mean + K.exp(z_log_var / 2) * epsilon
```

```python
inputs = keras.Input(shape=(trainset.shape[1],))
out = keras.layers.Dense(10, activation='relu')(inputs)

z_mean_layer = keras.layers.Dense(trainset.shape[1])
z_log_var_layer = keras.layers.Dense(trainset.shape[1])

z_mean = z_mean_layer(out)
z_log_var = z_log_var_layer(out)

z = keras.layers.Lambda(sampling)([z_mean, z_log_var])

mean_decoded = keras.layers.Dense(trainset.shape[1], activation='relu')(z)

vae = keras.Model(inputs, mean_decoded)

generate = keras.Model(inputs, [z_mean, z_log_var])
```

```python
generate.summary()
```

```python
xent_loss = trainset.shape[1] * keras.losses.MSE(K.reshape(inputs, (-1, trainset.shape[1])), mean_decoded)
kl_loss = - 0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
vae_loss = K.mean(xent_loss + kl_loss)

vae.add_loss(vae_loss)

vae.compile('adam')
```

```python
vae.fit(trainset, shuffle=True, epochs=3, batch_size=512)
```
